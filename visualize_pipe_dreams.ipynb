```python
import matplotlib.pyplot as plt
import sys
sys.path.insert(0, '/home/matthematics/schubmult/src')

from schubmult import Permutation, RCGraph
from schubmult.visualization import draw_pipe_dream, draw_pipe_dream_tikz
```

```python
# RC graph with a crossing in row 1 with right root (3,4)
# Right root (3,4) corresponds to simple reflection s_3, which is element 4
rc1 = RCGraph([(4,),(),(),()])
print("RC graph with crossing at right root (3,4) in row 1:")
print(rc1)
print(f"Permutation: {rc1.perm}")
draw_pipe_dream(rc1, top_labeled=True, flip_horizontal=False)
plt.show()
print(draw_pipe_dream_tikz(rc1, top_labeled=False, flip_horizontal=False))
```

## Example 1: Simple Permutation

```python
# Create RC graph for permutation (2, 1, 3)
perm = Permutation([2, 1, 3])
print(f"Permutation: {perm}")

# Get all RC graphs for this permutation
rc_graphs = list(RCGraph.all_rc_graphs(perm, 2))
print(f"Found {len(rc_graphs)} RC graph(s)")

for i, rc in enumerate(rc_graphs):
    print(f"\nRC Graph {i+1}:")
    print(rc)
    
    # Show both orientations
    draw_pipe_dream(rc, title=f"Pipe Dream for {perm} (RC Graph {i+1}) - Flipped (default)", flip_horizontal=True)
    plt.show()
    
    draw_pipe_dream(rc, title=f"Pipe Dream for {perm} (RC Graph {i+1}) - Not Flipped", flip_horizontal=False)
    plt.show()
```

## Example 2: Larger Permutation

```python
# Create RC graph for permutation (1, 3, 2)
perm = Permutation([1, 3, 2])
print(f"Permutation: {perm}")

rc_graphs = list(RCGraph.all_rc_graphs(perm, 2))
print(f"Found {len(rc_graphs)} RC graph(s)")

for i, rc in enumerate(rc_graphs):
    print(f"\nRC Graph {i+1}:")
    print(rc)
    draw_pipe_dream(rc, title=f"Pipe Dream for {perm} (RC Graph {i+1})")
    plt.show()
```

## Example 3: Before and After Insertion

Visualize how a pipe dream changes after Kogan-Kumar insertion.

```python
# Create an RC graph
perm = Permutation([1, 3, 2])
rc_graphs = list(RCGraph.all_rc_graphs(perm, 2))
rc = rc_graphs[0]

print("Before insertion:")
print(rc)
draw_pipe_dream(rc, title="Before Insertion")
plt.show()

# Apply Kogan-Kumar insertion
d = 2
r = 2
rc2, reflections = rc.kogan_kumar_insert(d, [r], return_reflections=True)

print(f"\nAfter inserting at descent={d}, row={r}:")
print(rc2)
print(f"Reflections added: {reflections}")
draw_pipe_dream(rc2, title=f"After Insertion (d={d}, r={r})")
plt.show()
```

## Example 4: Compare Multiple RC Graphs

```python
# Show all RC graphs for n=4
n = 4
all_perms = list(Permutation.all_permutations(n))

print(f"All permutations for n={n}:")
print(f"Total: {len(all_perms)} permutations\n")

for perm in all_perms:
    rc_graphs = list(RCGraph.all_rc_graphs(perm, n - 1))
    print(f"Permutation {perm}: {len(rc_graphs)} RC graph(s)")
    
    for i, rc in enumerate(rc_graphs):
        print(f"  RC Graph {i+1}: {rc}")
        draw_pipe_dream(rc, title=f"Perm {perm} - RC Graph {i+1}")
        plt.show()
    print()
```
